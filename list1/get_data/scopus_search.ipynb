```python
from pyscopus import Scopus
import pandas as pd
```

```python
# your scopus API key here
key = 'xxxxx'
scopus = Scopus(key)
```

```python
title = "cs_emps" #without .csv
f_emps = pd.read_csv('{}.csv'.format(title))
```

```python
ids = []

for index, row in f_emps.iterrows():
    uni = 'Wrocław University of Science and Technology'
    name = row['name']
    surname = row['surname']
    searchstr = 'AUTHLASTNAME({0}) and AUTHFIRST({1}) and AFFIL({2})'.format(surname, name, uni)
    print(name + ' ' + surname)
    try:
        author_result_df = scopus.search_author(searchstr)
        ids.append(author_result_df['author_id'].iloc[0])   
    except:
        
        ids.append(None) 
        pass
        
```

```python
ids
```

```python
temp = f_emps.copy(deep=True)
```

```python
temp['scopus_id'] = ids
temp['scopus_id'] = temp['scopus_id'].apply(str)
```

```python
emps = temp
emps
```

```python
co_lst = []
for emp_id in emps["scopus_id"]:
    if emp_id == "None":
        co_lst.append("None")
        continue
    pub_df = scopus.search_author_publication(emp_id)
    co_lst.append(list(pub_df["authors"]))
```

```python
assc_df = emps.copy(deep=True)
assc_df["pubs"] = co_lst
assc_df
```

```python
assc_df.to_csv("{}_scopus.csv".format(title))
```
